```python
%env FLASK_APP=abacus.py
%env SQLITE = "True"
%env DATABASE_NAME = abacus
%env SQLITE_DATABASE_PATH=/abacus.db
%env DEBUG = 1

from flask_sqlalchemy import SQLAlchemy
from app import connex_app,app, db
import connexion
from app.models import Dimension
from config import basedir
import pandas as pd
import numpy as np
```

```python
tags = pd.read_csv("tags.csv")
tags = tags.replace({np.nan: None})
```

```python
dim_list = []
for _, row in tags.iterrows():
    dim_list.append(
        Dimension(
        name=row["name"],
        primary_tag=row["primary_tag"],
        secondary_tag=row["secondary_tag"],
        data_group=row["data_group"],
        )
    )    
    with app.app_context():
        db.create_all()
```

```python
with app.app_context():
    db.session.add_all(dim_list)
    db.session.commit()
```
